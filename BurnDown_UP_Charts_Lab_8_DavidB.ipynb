```python
myData = pd.read_csv('/content/drive/My Drive/ISU/IT 262/BDU.csv')
```

```python
myData.shape
myData.head(2)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(40,10))


ax1.set_xlabel('days')
ax2.set_xlabel('days')

ax1.plot(myData.Cideal, color = 'grey', label = 'work completed - ideal')
ax3 = ax1.twinx()
ax3.plot(myData.Cactual, color = 'blue', label = 'work completed - actual')

ax2.plot(myData.Lideal, color = 'grey', label = 'work completed - ideal')
ax4 = ax2.twinx()
ax4.plot(myData.Lactual, color = 'blue', label = 'work completed - actual')

ax1.set_title('burn up')
ax2.set_title('burn down')

h1, l1 = ax1.get_legend_handles_labels()
h3, l3 = ax3.get_legend_handles_labels()
ax3.legend(h1+h3,l1+l3,loc=2,prop={'size':20})
h2, l2 = ax2.get_legend_handles_labels()
h4, l4 = ax4.get_legend_handles_labels()
ax4.legend(h1+h3,l1+l3,loc=1,prop={'size':20})





plt.show()
```
